```python
lists=['Vehicle_driver_relation', 'Work_of_casuality', 'Fitness_of_casuality','Day_of_week','Casualty_severity','Time','Sex_of_driver','Educational_level','Defect_of_vehicle','Owner_of_vehicle','Service_year_of_vehicle', 'Road_surface_type','Sex_of_casualty']
dataframe.drop(columns = lists, inplace=True)
dataframe
```

```python
#Replacing the null values with mode values.
dataframe['Driving_experience'].fillna(dataframe['Driving_experience'].mode()[0], inplace=True)
dataframe['Age_band_of_driver'].fillna(dataframe['Age_band_of_driver'].mode()[0], inplace=True)
dataframe['Type_of_vehicle'].fillna(dataframe['Type_of_vehicle'].mode()[0], inplace=True)
dataframe['Area_accident_occured'].fillna(dataframe['Area_accident_occured'].mode()[0], inplace=True)
dataframe['Road_allignment'].fillna(dataframe['Road_allignment'].mode()[0], inplace=True)
dataframe['Type_of_collision'].fillna(dataframe['Type_of_collision'].mode()[0], inplace=True)
dataframe['Vehicle_movement'].fillna(dataframe['Vehicle_movement'].mode()[0], inplace=True)
dataframe['Lanes_or_Medians'].fillna(dataframe['Lanes_or_Medians'].mode()[0], inplace=True)
dataframe['Types_of_Junction'].fillna(dataframe['Types_of_Junction'].mode()[0], inplace=True)
```

```python
dataframe.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df=dataframe.apply(LE.fit_transform)                              #categorical values to integers
```

```python
plt.figure(figsize=[25,15])
sns.heatmap(df.corr(),annot=True)
```

```python
#Spliting the data
from sklearn.model_selection import train_test_split
x = df.drop('Accident_severity', axis=1)
y = df['Accident_severity']

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=42)
print(xtrain.shape, xtest.shape, ytrain.shape, ytest.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(xtrain,ytrain)
```

```python
y_pred = lr.predict(xtest)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy =  accuracy_score(ytest,y_pred)
print('Accuracy score of Logistic Regression:{0:.2f}%'.format(accuracy*100))
print(classification_report(ytest,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
decisiontree = DecisionTreeClassifier()
decisiontree.fit(xtrain,ytrain)
```

```python
y_pred = decisiontree.predict(xtest)
y_pred
```

```python
from sklearn.metrics import accuracy_score
decisionaccuracy =  accuracy_score(ytest,y_pred)
print('Accuracy score of Decision Tree:{0:.2f}%'.format(decisionaccuracy*100))
print(classification_report(ytest,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
RandomTree = RandomForestClassifier()
RandomTree.fit(xtrain,ytrain)
```

```python
y_pred = RandomTree.predict(xtest)
y_pred
```

```python
from sklearn.metrics import accuracy_score
RandomForestaccuracy =  accuracy_score(ytest,y_pred)
print('Accuracy score of Random Forest:{0:.2f}%'.format(RandomForestaccuracy*100))
print(classification_report(ytest,y_pred))
```

```python
from sklearn.linear_model import LinearRegression
linearR = LinearRegression()
linearR.fit(xtrain,ytrain)
```

```python
y_pred = linearR.predict(xtest)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
linearmse = mean_squared_error(ytest,y_pred)
print(linearmse)
```
