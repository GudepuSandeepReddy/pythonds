```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,8,12,6,14]
plt.plot(x,y)
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('X-axis label')
plt.show()
```

```python
#creating dataframes
import pandas as pd
df = pd.DataFrame({
    "column1":[1,4,8,7,9],
    "column2":['a','column','with','a','string'],
    "column3":[1.23,23.5,45.6,32.17,89.45],
    "column4":[True,False,True,False,True]
})
print(df)
```

```python
#creating series
#printing index values
import pandas as pd
s=pd.Series([1,2,3,4])
print(s)
print(s.values)
print(s.index.values)
print(s.index)
```

```python
#assigning index names
import pandas as pd
s=pd.Series([1,2.3,'a'], index=["india","china","usa"])
s.name='something'
s.index.name='country'
print(s['india'])
```

```python
#droping entry in series
import pandas as pd
s=pd.Series([1,2.3,'a'], index=["india","china","usa"])
s.name='something'
s.index.name='country'
s=s.drop('india')
print(s)
```
