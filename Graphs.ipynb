```python
#line plot
import matplotlib.pyplot as plt
x=[1,2,5,3,8,8,6]
y=[3,5,2,1,23,7,34]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#bar graph
import seaborn as sns
import matplotlib.pyplot as plt

categories=['a','b','c','d','e']
values=[25,35,45,55,65]

sns.barplot(x=categories,y=values)
plt.xlabel('Categories')
plt.ylabel('values')
plt.show()
```

```python
#scatter plot
import plotly.express as px
x=[1,3,3,14,4]
y=[12,4,2,5,6]
fig=px.scatter(x=x,y=y)
fig.show()
```
